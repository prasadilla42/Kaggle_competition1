```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_test=pd.read_csv('test.csv')
```

```python
df_test.head()
```

```python
df_test.shape
```

```python
for feature in df_test.columns:
    if (df_test[feature].isnull().sum()!=0):
        print("Null values of the "+feature+df_test[feature].isnull().sum())
print("No null values found")
```

```python
categorical_features=[feature for feature in df_test.columns if df_test[feature].dtype=='O']
```

```python
#Get dummy variables for all the categorical variables
def one_hot_top(df,variable,top_x_labels):
#function to create dummy variables
    for label in top_x_labels:
        df[variable+'_'+label]=np.where(df[variable]==label,1,0)
#Create a new dataframe only with categorical variables.
data=pd.read_csv('test.csv',usecols=categorical_features)
```

```python
for feature in data.columns:
    top_10=[x for x in data[feature].value_counts().sort_values(ascending=False).head(10).index]
    one_hot_top(data,feature,top_10)
```

```python
data.drop(categorical_features,axis=1,inplace=True)
```

```python
df_test_new=pd.concat([df_test,data],axis=1)
```

```python
df_test_new.drop(categorical_features,axis=1,inplace=True)
```

```python
df_test_new.head()
```

```python
X_test=df_test_new
```

```python
X_test.head()
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel
```

```python
y_train=pd.read_csv('y_train.csv')
```

```python
feature_sel_model = SelectFromModel(Lasso(alpha=0.005, random_state=0)) # remember to set the seed, the random state in this function
feature_sel_model.fit(X_test,y_train)
```

```python
# let's print the number of total and selected features

# this is how we can make a list of the selected features
selected_feat = X_test.columns[(feature_sel_model.get_support())]

# let's print some stats
print('total features: {}'.format((X_test.shape[1])))
print('selected features: {}'.format(len(selected_feat)))
```

```python
X_test.to_csv('X_test.csv')
```
